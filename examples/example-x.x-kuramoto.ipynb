```python
import numpy as np

from neurolib.models.kuramoto.model import KuramotoModel

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
model = KuramotoModel()
model.params['duration'] = 20
model.params['dt'] = 0.01
model.params['N'] = 8
model.params['lengthMat'] = np.zeros((model.params['N'], model.params['N']))
model.params['Cmat'] = np.ones((model.params['N'], model.params['N']))
model.params['k'] = 2
model.params['thetas_init'] = np.random.uniform(low=0, high=2*np.pi, size=(model.params['N'], 1))
model.params['theta_ou'] = np.zeros((model.params['N'],))
model.params['omega'] = np.random.normal(loc=np.pi, scale=np.pi, size=(model.params['N'],))

model.params['sigma_ou'] = 1
```

```python
model.run()
```

```python
theta = model['theta']
```

```python
plt.plot(model.t, theta.T)
plt.show()
```
